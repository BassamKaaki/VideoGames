```
########## XGBOOST ANALYSIS ##########
```

```
df_boost = vgsales
```

```
df_boost
```

```
#Convert categorical variables to numerical
from sklearn import model_selection, preprocessing
for c in df_boost.columns:
  if df_boost[c].dtype == 'object':
    lbl = preprocessing.LabelEncoder()
    lbl.fit(list(df_boost[c].values))
    df_boost[c] = lbl.transform(list(df_boost[c].values))
```

```
df_boost
```

```
#Checking plots to see linearity
plt.figure(figsize = (10,5))
sns.regplot(x = 'Genre', y = 'Global_Sales', data = df_boost)
```

```
#Checking plots to see linearity
plt.figure(figsize = (10,5))
sns.regplot(x = 'Platform_Type', y = 'Global_Sales', data = df_boost)
```

```
#Checking plots to see linearity
plt.figure(figsize = (10,5))
sns.regplot(x = 'Rank', y = 'Global_Sales', data = df_boost)
```

```
#Train - Test - Split
from sklearn.model_selection import train_test_split
label = df_boost.pop('Global_Sales')
data_train, data_test, label_train, label_test = train_test_split(df_boost, label, test_size = 0.2, random_state = 9)
```

```
data_train.shape, data_test.shape
```

```
########## Extremem Gradient Boosting - check on most important feature ##########
import xgboost as xgb
from sklearn.model_selection import KFold, train_test_split, GridSearchCV
```

```
### Define Parameters ###
xgb_params = {
    'eta': 0.05,
    'max_depth': 5,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'silent': 1
}
```

```
dtrain = xgb.DMatrix(data_train, label_train)
```

```
#Cross Validation of XBoost
cv_output = xgb.cv(xgb_params, dtrain, num_boost_round = 1000, early_stopping_rounds = 20,
                   verbose_eval = 50, show_stdv = False)
```

```
num_boost_rounds = len(cv_output)
model = xgb.train(dict(xgb_params, silent = 0), dtrain, num_boost_round = num_boost_rounds)
```

```
fig, ax = plt.subplots(1, 1, figsize = (8, 10))
xgb.plot_importance(model, max_num_features = 50, height = 0.5, ax = ax, color = 'cyan')
#Shows that company rankings are very important as most of trees gave vote to the rank feature.
```
